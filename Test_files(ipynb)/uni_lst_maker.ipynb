```python
def get_dropdown_options(lst):
    uni_lst_temp = []
    for val in lst:
        uni_lst_temp.extend(val.replace("[", '').replace("'","").split(', '))
    out = []
    for name in uni_lst_temp:
        if len(name)<3:
            continue
        if '@' in name:
            out.append(name[name.find('@'):])
        else:
            out.append(name)
    return list(set(out))
%timeit get_dropdown_options(df1['affilliations'].values.tolist())
```
