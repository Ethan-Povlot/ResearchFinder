````python
lst =df['affiliations'].values.tolist()

flat_list = []
for sublist in lst:
    try:
        for item in sublist:
            if '@' in item:
                item = item[item.find('@'):]
                if '.edu' in item:
                    item = item[:item.find('.edu')+4]
                if item[-1] == '.':
                    item = item[:-1]
                item = item.strip()
            if item in ['Central', 'The U']:
                continue
            if len(item)>3:
                flat_list.append(item)
    except:
        pass
output = []
counts = Counter(flat_list)
x = list(zip(list(counts.keys()), list(counts.values())))
x.sort(key=lambda pair: pair[1], reverse=True)

for i in x:
    output.append(i[0]+'\t\t\t('+str(i[1])+')')
````

````python
def get_dropdown_options(lst):
    global output
    uni_lst_temp = []
    for val in lst:
        for uni in str(val).replace("[", '').replace("]", '').replace("'","").replace(";", ',').split(', '):
            uni_lst_temp.append(uni.strip())
    flat_list = []
    for item in uni_lst_temp:
        try:
            if '@' in item:
                item = item[item.find('@'):]
                if '.edu' in item:
                    item = item[:item.find('.edu')+4]
                if item[-1] == '.':
                    item = item[:-1]
                item = item.strip()
            if str(item) in ['Central', 'The U', 'nan']:
                continue
            if len(item)>3:
                flat_list.append(item)
        except:
            pass
    output = []
    counts = Counter(flat_list)
    x = list(zip(list(counts.keys()), list(counts.values())))
    x.sort(key=lambda pair: pair[1], reverse=True)

    for i in x:
        output.append(i[0]+'\t\t\t('+str(i[1])+')')

    known_schools= ['@emory.edu', '@gatech.edu', 'Georgia Institute of Technology', 'gatech']
    for names in known_schools:
        if names in output:
            output.remove(names)
    print(output)
    return ['All']+output
uni_init_lst = get_dropdown_options(df['affiliations'].values.tolist())#['All']#
````
